3. Data Reduction: Pick two variables that could be simplified by binning or bucketing, or by normalization or standardization. Then perform the operation(s) you have chosen. Insert a text section into your notebook and explain in no less than 3 sentences what benefit this may provide for any later analysis. (3 points)

```python
olympic.Weight.describe()
```

```python
Weight_bins = np.linspace(olympic.Weight.min(),olympic.Weight.max(),4)
labels = ['small', 'medium', 'high']
olympic3 = olympic
olympic3['Weight_bins'] = pd.cut(olympic3['Weight'], bins=Weight_bins, labels=labels, include_lowest=True)
olympic3.groupby(['Weight_bins']).count()
```

```python
olympic.Age.describe()
```

```python
Age_bins = np.linspace(olympic.Age.min(),olympic.Age.max(),4)
labels = ['Young', 'medium', 'Old']
olympic3 = olympic
olympic3['Age_bins'] = pd.cut(olympic3['Age'], bins=Age_bins, labels=labels, include_lowest=True)
olympic3.groupby(['Age_bins']).count()
```

The chosen variables are both the Weight and the Age. In later analysis bining will help understand which player Weight is the most ideal depending on other variables such as the sport involved and even their respective countries. The Age bining will also be ideal in analyzing the best Age to star in particular Sports.
